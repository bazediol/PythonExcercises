```python
import re
protocols = {'L':'local','C':'connected','S':'static','R':'RIP','M':'mobile','B':'BGP','D':'EIGRP',
             'EX':'external','O':'OSPF','IA':'inter area','N1':'NSSA external type 1',
             'N2':'NSSA external type 2','E2':'external type 2','E':'EGP','i':'IS-IS',
             'L1':'level-1','L2':'level 2','is':'inter area','*':'candidate default',
             'U':'per-user static route','o':'ODR','P':'periodic downloaded static route','H':'NHRP','l':'LISP'}
route_pattern = re.compile ('(\w*) *(.*) ((\d{1,3}\.){3}\d{1,3}) \[(\d*\/\d*)\] via ((\d{1,3}\.){3}\d{1,3}), (\d*:\d*:\d*), (.*)')
route_connected = re.compile ('(C) ((\d{1,3}\.){3}\d{1,3}) ((\d{1,3}\.){3}\d{1,3}) .*, (.*)')
def detect_protocol(reduction1, reduction2):
    if reduction1 and reduction2 in protocols:
        return '{} - {}'.format(protocols[reduction1],protocols[reduction2])
    elif reduction1 in protocols:
        return protocols[reduction1]
    else: 
        return ''
def route_output(line):
    print ('\n\n{:<20} {}\n{:<20} {}\n{:<20} {}\n{:<20} {}\n{:<20} {}\n{:<20} {}'.format('Protocol:',detect_protocol(line[0],line[1]),'nPrefix:', line[2],'AD/Metric:', line[4], 'Nex-Hop:', line[5], 'Last Update:', line[7],'Outbound interface:', line[8]))
def connected_output(line):
    print ('\n\n{} {}\n{} {}\n{} {}\n{} {}'.format('Type:',detect_protocol(line[0], ''),'Prefix:',line[1],'Nex-Hop:', line[3],'Outbound interface',line[5]))

file = open('ShowIpRoute.txt', 'r')
if (file.closed == False):
    print('\nFile ShowIpRoute.txt is opened now')
else:
    print('\nFile wasn\'t opened :(')
file_content = file.read()
file.close()
if (file.closed == True):
    print('\nFile commands.txt is closed now')
else:
    print ('\nUnable to close the file')
routes = re.findall(route_pattern,file_content)
connected_routes = re.findall (route_connected,file_content)
for line in connected_routes:
    connected_output(line)
for line in routes:
    route_output(line)
```
